## Partie théorique

### T1)


$$
    \newcommand{\Pxy}{\mathbb{P}(X_1 = y | X_0 = x)}
    \newcommand{\proba}[1]{\mathbb{P}(#1)} 
$$
On définit $P(x,y)$ par 

$$
\begin{equation}
P(x,y) = \Pxy
\end{equation}
$$

Observons que $P(x,y)$ est bien définie $\forall x, y \in E$ car il est une probabilité conditionnelle, qui est bien définie.

Premièrement, supposons $x \neq y$. On utilise la formule des probabilités totales pour les deux cas $X_1 = Y_1$ ou $X_1 = X_0$:

$$
\begin{align}
P(x,y) &= \proba{Y_1 = y | X_0 = x, U_1 < R(x,y)}\proba{U_1 < R(x,y)}+\proba{X_0 = y | X_0 = x, U_1 \geq R(x,y)}\proba{U_1 \geq R(x,y)}\\
&= \proba{Y_1 = y | X_0 = x, U_1 < R(x,y)}\proba{U_1 < R(x,y)}\\
&= \frac{\proba{Y_1 = y, X_0 = x, U_1 < R(x,y)}\proba{U_1 < R(x,y)}}{\proba{X_0 = x, U_1 < R(x,y)}}\\
&= \proba{Y_1 = y} \proba{U_1 < R(x,y)}\\
&= q(y) R(x,y)
\end{align}
$$

Dans le cas où $x = y$, il suffit d'utiliser la relation

$$
\begin{equation}
\Pxy +\proba{X_1 \neq y | X_0 = x} = 1
\end{equation}
$$

Et on obtient donc

$$
\begin{align}
\Pxy &= 1 - \proba{X_1 \neq y | X_0 = x}\\
&= 1 - \sum_{z \neq y} \proba{X_1 = z | X_0 = x}
\end{align}
$$

Pour $x \neq y$, il est évident que $P(x,y)>0$. Pour $x=y$, 

$$
\begin{align}
P(x,y) &= 1 - \sum_{z \neq y} \proba{X_1 = z | X_0 = x}\\
&= 1 - \sum_{z \neq y} q(z) \min{\left(\frac{\pi(z)q(x)}{\pi(x)q(z)},1\right)}\\
&= 1 - \sum_{z \neq y} \min{\left(\frac{\pi(z)q(x)}{\pi(x)},q(z) \right)}\\
&\geq 1 - \sum_{z \neq y} q(z)\\
&= 1 - (1 - q(y)) = q(y) > 0
\end{align}
$$

### T2)

L'identité est triviale pour $x = y$. Supposons $x \neq y$ :

$$\begin{align}
\pi(x)P(x,y) &= \pi(x)q(y)R(x,y)\\
&= \min{\left \{ \pi(y)q(x),\pi(x)q(y) \right \}}\\
&= \pi(y)q(x)\min{\left(1,\frac{\pi(x)q(y)}{\pi(y)q(x)} \right)}\\
&= \pi(y)q(x)R(y,x) \\
&= \pi(y) P(y,x)
\end{align}$$

### T3) 

Pour écrire une formule pour $\proba{X_1 = y}$, on utilise la formule des probabilités totales :

$$\begin{align}
\proba{X_1 = y} &= \sum_{x \in E} \Pxy \proba{X_0 = x}\\
& = \sum_{x \in E} \mu (x) P(x,y)
\end{align}$$

Pour trouver la rélation pour $n$ quelconque, on commence par la formule des probabilités totales :
$$\begin{equation}
\proba{X_{n+1}=y} = \sum_{x \in E} \proba{X_{n+1} = y | X_n = x} \proba{X_n = x}
\end{equation}$$

Alors, pour $X_n = x$ donné, la définition de $X_{n+1}$ ne dépend que des variables $Y_{n+1}$ et $U_{n+1}$. En utilisant le fait que $(U_{n+1},Y_{n+1})$ est indépendante de $(U_n, Y_n)$ et le fait que $Y_n = Y_{n+1}$, $U_n = U_{n+1}$ **en loi**, alors on déduit que 

$$\begin{equation}
 \proba{X_{n+1} = y | X_n = x} = \proba{X_{n} = y | X_{n-1} = x}
\end{equation}$$

Et par récurrence cette probabilité est bien $P(x,y)$, d'où

$$\begin{equation}
\proba{X_{n+1}=y} = \sum_{x \in E} P(x,y) \proba{X_n = x}
\end{equation}$$

### T4)

Si $\mu = \pi$, on utilise la relation trouvé em **T3)** pour écrire

$$\begin{align}
\pi_1 (y) & = \sum_{x \in E} \pi (x) P(x,y)\\
& = \sum_{x \in E} \pi (y) P(y,x) \hspace{5mm} \text{(par $\mathbf{T2}$)}\\
&= \pi (y) \sum_{x \in E} \proba{X_1 = x | X_0 = y}\\
&= \pi (y)
\end{align}$$

Ensuite, on montre le résultat par récurrence : supposons $\pi_n = \pi$. On utilise le deuxième résultat de **T3)** :

$$\begin{align}
\pi_{n+1}(y) & = \sum_{x \in E} \pi (x) P(x,y)\\
& = \sum_{x \in E} \pi (y) P(y,x) \hspace{5mm} \text{(par $\mathbf{T2}$)}\\
&= \pi (y) \sum_{x \in E} \proba{X_1 = x | X_0 = y}\\
&= \pi (y)
\end{align}$$


### T5) 

On observe que

$$\begin{align}
P(x,y)-K^{-1}\pi (y) &= q(y)\min{\left(1,\frac{\pi(y)q(x)}{\pi(x)q(y)} \right)} -K^{-1}\pi (y)\\
&= \pi(y)\left(\min{\left(\frac{q(y)}{\pi(y)},\frac{q(x)}{\pi(x)} \right)} -\inf_{z \in E} \frac{q(z)}{\pi(z)}\right)\\
&\geq \pi(y) \left(\inf_{z \in E} \frac{q(z)}{\pi(z)}-\inf_{z \in E} \frac{q(z)}{\pi(z)} \right) = 0
\end{align}$$

### T6)

Par la définition de $d_{VT}$,

$$
\begin{align}
    d_{VT}(\pi_{n+1}, \pi) &= \frac{1}{2}\sum_{y}\left|\pi_{n+1}(y)-\pi(y)\right|\\
    &= \frac{1}{2}\sum_{y}\left|\sum_{x}\pi_n(x)P(x,y)-\sum_{x}\pi(x)P(x,y)\right| \hspace{5mm} \text{(par $\mathbf{T4}$)}\\ 
    &= \frac{1}{2}\sum_{y}\left|\sum_{x}(\pi_n(x) - \pi(x) )P(x,y)\right|
\end{align}
$$

Ensuite, on utilise le fait que $\sum_x \pi_n(x) = \sum_x \pi(x) = 1$, et donc $\sum_x (\pi_n(x)-\pi(x))=0$. En particulier, $K^{-1}\pi(y)\sum_x (\pi_n(x)-\pi(x))=0$, et finalement

$$
\begin{align}
    d_{VT}(\pi_{n+1}, \pi) &= \frac{1}{2}\sum_{y}\left|\sum_{x}(\pi_n(x) - \pi(x) )P(x,y) - K^{-1}\pi(y)\sum_x (\pi_n(x)-\pi(x))\right|\\
    &= \frac{1}{2}\sum_{y}\left|\sum_{x}(\pi_n(x) - \pi(x) )(P(x,y) - K^{-1}\pi(y))\right|
\end{align}
$$

### T7) 

D'après **T5)**, c'est facile de voir que

$$\begin{align} 
P(x,y)-K^{-1}\pi (y) &= q(y)\min{\left(1,\frac{\pi(y)q(x)}{\pi(x)q(y)} \right)} -K^{-1}\pi (y)\\
&\leq q(y)\min{\left(1,\frac{\pi(y)q(x)}{\pi(x)q(y)} \right)}\\
&\leq q(y)
\end{align}$$

Ainsi,

$$\begin{align}
\frac{1}{2}\sum_{y \in E} \sum_{x \in E} |(\pi_n(x)-\pi_(x))(P(x,y)-K^{-1}\pi (y))| &\leq \frac{1}{2}\sum_{y \in E} \sum_{x \in E} |(\pi_n(x)-\pi(x))q(y)| \\
&= \sum_{y \in E} q(y) \frac{1}{2}\sum_{x \in E} |\pi_n(x)-\pi_(x)|\\
&= d_{VT}(\pi_n,\pi) \sum_{y \in E} q(y)\\
&= d_{VT}(\pi_n,\pi)
\end{align}$$

Par récurrence, la série ci-dessus converge absoluement $\forall n \in \mathbb{N}$. Cela nous permet d'appliquer le théorème de Fubini pour commuter les sommes en $x$ et en $y$ :

$$\begin{align}
d_{VT}(\pi_{n+1},\pi) &\leq \frac{1}{2}\sum_{y \in E} \sum_{x \in E} |(\pi_n(x)-\pi(x))(P(x,y)-K^{-1}\pi (y))| \\
&= \frac{1}{2}\sum_{x \in E} \sum_{y \in E} |(\pi_n(x)-\pi(x))(P(x,y)-K^{-1}\pi (y))|\\
&= \frac{1}{2}\sum_{x \in E}|(\pi_n(x)-\pi(x))|\sum_{y \in E}(P(x,y)-K^{-1}\pi (y))\\
&= \frac{1}{2}\sum_{x \in E}|(\pi_n(x)-\pi(x))|\left(\sum_{y \in E}\Pxy-K^{-1}\sum_{y \in E}\pi (y)\right)\\
&= \frac{1}{2}\sum_{x \in E}|(\pi_n(x)-\pi(x))|(1-K^{-1})\\
&= d_{VT}(\pi_n,\pi)(1-K^{-1})
\end{align}$$

## L'algorithme

### S1)

On commence l'implementation de l'agorithme. Premièrement, on fait l'`import` des libraries que l'on utilisera :

```python
import math
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
import time
```

```python
%matplotlib inline
%config InlineBackend.figure_format ='retina' # pour avoir une résolution double sur un écran Retina (macbook)
```

Maintenant, on définit la fonction $\pi(n)$ non-normalisée :

```python
def function_pi(n, lambda_p):
    return math.pow(lambda_p,n)/(math.log(n+2)*math.factorial(n))
```

Pour computer $Z_\lambda$, on peut calculer la somme $S(n) = \sum_{i=0}^n \pi(n)$ jusqu'à le critère d'arret

$$\begin{equation}
\left|\frac{S(n+1)-S(n)}{S(n)}\right| \leq \varepsilon
\end{equation}$$

Où $\varepsilon$ est très petit. Ce critère est valide car, par la définition de $Z_\lambda$, on peut facilement observer que $Z_{\lambda, n} = S(n)$, et donc la convergence vers $Z_{\lambda}$ est égale à celle de $S(n) \rightarrow S_{\infty}.$

```python
def normal_constant_crit(lambda_p, stop_crit):
    s_cur = function_pi(0,lambda_p)
    i = 0
    while(math.fabs(function_pi(i+1,lambda_p)/s_cur) > stop_crit):
        i = i+1
        s_cur = s_cur + function_pi(i, lambda_p)
    ans = [s_cur, i]
    return ans
```

Calculons la valeur de $Z_\lambda$ lorsque $\lambda = 0.5, 1, 2$. On choisira $\varepsilon = 10^{-9}$ :

```python
stop_crit = math.pow(10, -9)
Z = [normal_constant_crit(0.5, stop_crit), normal_constant_crit(1, stop_crit), normal_constant_crit(2, stop_crit)]

print(Z)
```

On peut observer que le nombre d'iterations requis pour calculer la valeur de $Z_\lambda$ est très petit, vu que, en pratique, il est rare d'observer $X \geq 15$ pour une variable aleatorie $X \sim \pi(\lambda)$ si $\lambda$ est assez petit. Pour faciliter l'utilisation de $\pi(x)$ dans les questions suivantes, on définit

```python
def function_pi_norm(n, lambda_p):
    Z = normal_constant_crit(lambda_p, math.pow(10, -9))
    return function_pi(n, lambda_p)/Z[0]
```

Pour lancer l'algorithme, on n'a pas besoin de connaître la valeur de $Z_\lambda$, car il disparaît losrqu'on calcule le quotient $\pi(x)/\pi(y)$ dans la règle d'acceptation. Toutefois, il sera important pour l'item **S2)**.

Maintenant, on définit les fonctions qui nous aideront à faire les calculs dans l'algorithme :

```python
def function_poisson(n, lambda_p):
    return (math.exp(-lambda_p)*math.pow(lambda_p,n))/math.factorial(n)
```

```python
def acceptation(x, y, lambda_p):
    return function_poisson(x, lambda_p)*function_pi(y, lambda_p)/(function_poisson(y, lambda_p)*function_pi(x, lambda_p))
```

Initialisation des parametres. Ici, `n_sample` répresente le parametre $L$ :

```python
n_sample = 1000
```

Finalement, l'implementation de l'algorithme. La fonction `xn_estimator` ne correspond pas à l'estimateur suggeré par l'énoncé. En effet, elle correspond au vecteur de $L$ tirages de $\pi_N$. L'estimateur est défini proprement dans la fonction `hist_plot`.

```python
def metropolis(N, lambda_p):
    x = np.zeros(N+1, dtype=int)
    y = int(npr.poisson(lambda_p))
    x[0] = y
    
    for i in range(N):
        y = int(npr.poisson(lambda_p))
        u = np.random.uniform()
        #If acceptation > 1, then u will surely be smaller than acceptation. 
        if u < acceptation(x[i],y,lambda_p):
            x[i+1] = y
        else:
            x[i+1] = x[i]
    return x
```

```python
def xn_estimator(N, lambda_p, n_sample):
    xn = np.zeros(n_sample, dtype=int)
    for i in range(n_sample):
        xn[i] = metropolis(N, lambda_p)[N]
    return xn
```

```python
def hist_plot(array, N, n_sample, lambda_p, ax):
    x_r = np.arange(15)
    p_sample = np.histogram(array, bins = np.append(x_r, np.inf))[0] / n_sample
    ax.bar(x = x_r, height = p_sample, edgecolor = "black", color = "blue", \
       width = 0.5, alpha = 0.2)
    ax.set(title = f"$\lambda = {lambda_p}$, $N = {N} $, $L = {n_sample}$",
           xlabel = "État $\pi_N^L(n)$",
           ylabel = "Probabilité empirique")
    ax.grid(axis = 'y')
    return ax
```

```python
def plot_histogram_estimators(lambda_p):
    N = [5, 50, 100, 1000]
    xtest = []
    for i in range(len(N)):
        xtest.append(xn_estimator(N[i], lambda_p, n_sample));

    fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize=(16,12))
    for i, ax in zip(range(len(xtest)), axs.flat):
        hist_plot(xtest[i], N[i], n_sample, lambda_p, ax)
```

On trace les histogrammes corréspondants à la valeur $\lambda = 0.5$ :

```python
plot_histogram_estimators(0.5)
```

De même pour $\lambda = 1$ :

```python
plot_histogram_estimators(1)
```

Finalement, pour $\lambda = 2$ :

```python
plot_histogram_estimators(2)
```

Notons qu'en utilisant l'estimateur

$$\begin{equation}
\pi_N^L(m) = \frac{1}{L}\sum_{l=1}^L \mathbf{1}_m(X_N^l)
\end{equation} $$
Il faut lancer l'algorithme $L$ fois, d'où on obtient une complexité de O($NL$). On pourrait, par contre, utiliser l'estimateur
$$\begin{equation}
\pi_N^{' L}(m) = \frac{1}{L}\sum_{l=N-L}^N \mathbf{1}_m(X_l)
\end{equation} $$
En supposant que por $1 \ll L \leq N$, on obtiendrait un résultat proche de celui déjà fait ci-dessus, mais cette fois-ci avec complexité O($N$). On peut comparer directement les temps de calcul pour chaque estimateur :

```python
def metropolis_v2(N, lambda_p, n_size):
    pts = np.zeros(n_size)
    x = np.zeros(N)
    y = npr.poisson(lambda_p)
    x[0] = y
    j = 0
    
    for i in range(N):
            y = npr.poisson(lambda_p)
            u = np.random.uniform()
            #If acceptation > 1, then u will surely be smaller than acceptation. 
            if u < acceptation(x[i-1],y,lambda_p):
                x[i] = y
            else:
                x[i] = x[i-1]
            if i>N/2:
                pts[j] = x[i]
                j = j+1
                
    return pts
```

```python
lambdap = 1
N = [2000, 1000]
tick1 = time.time()
x_pi = metropolis_v2(N[0], lambdap, n_sample)
tick1 = time.time() - tick1

tick2 = time.time()
x_pi_prime = xn_estimator(N[1], lambdap, n_sample)
tick2 = time.time() - tick2
print(tick2/tick1)
```

On peut tracer les graphes pour chaque méthode. Ici, $L$ répresente la taille de chaque echantillon.

```python
estimator_plot = [x_pi, x_pi_prime]
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(16,6))
for i, ax in zip(range(len(estimator_plot)), axs.flat):
    hist_plot(estimator_plot[i], N[i], n_sample, lambdap, ax)
```

On peut aussi calculer leurs distances $d_{VT}$ par rapport à la loi $\pi$ :

```python
def dist(array, trunc, lambda_p):
    distance = 0
    u = trunc
    if len(array) < trunc:
        u = len(array)
    for i in range(u):
        distance = distance + math.fabs(function_pi_norm(i, lambda_p) - array[i])
    return distance/2
```

```python
x_pi = x_pi.astype(int)
x_pi_prime = x_pi_prime.astype(int)
x_pi = np.bincount(x_pi, minlength = 20)/n_sample
x_pi_prime = np.bincount(x_pi_prime, minlength = 20)/n_sample
print("Distance VT nouvelle methode : ", dist(x_pi, 20, 1), ", Distance VT methode normale : ",dist(x_pi_prime, 20, 1))
```

On conclut que leurs distances par rapport à la loi de $\pi$, en utilisant l'estimateur de **S2)**, sont proches, mais la courbe de l'histogramme à droite est plus précise, ce qui est d'accord avec le fait que son coût computationnel est plus élevé.

### S2)

Le calcul de $K$ peut être fait explicitement, vu que, pour $q(x)$ loi de Poisson et $\pi(x)$ définie en **S1)**,

$$
\begin{equation}
\frac{1}{K} = \inf_{x \in \mathbb{N}} \frac{q(x)}{\pi(x)} = \inf_{x \in \mathbb{N}} Z_\lambda e^{-\lambda} log(x+2) = Z_\lambda e^{-\lambda} \log{2}
\end{equation} 
$$


```python
def K(lambda_p, Z_lambda):
    return math.exp(lambda_p)/(Z_lambda*math.log(2))

K = [K(0.5,Z[0][0]),K(1,Z[1][0]),K(2,Z[2][0])]
print(K)
```

Pour calculer $\pi_N^L(m)$, on peut utiliser la fonction `np.bincount` :

```python
lambdap = [0.5, 1, 2]
N = [5, 50, 100, 1000]
distance = np.zeros((len(lambdap),len(N)))
for i in range(len(lambdap)):
    xs2 = []
    for j in range(len(N)):
        xs2.append(xn_estimator(N[j], lambdap[i], n_sample))
    for j in range(len(N)):
        xs2[j] = np.bincount(xs2[j], minlength = 20)/n_sample
        distance[i][j] = dist(xs2[j], 20, lambdap[i])
    for k in range(len(N)-1):
        print("Teste pour = ",N[k+1],",",N[k]," lambda = ", lambdap[i], " : ", distance[i][k+1]/distance[i][k] <= math.pow(1-1/K[i],N[k+1]-N[k]))
```

On observe que l'algorithme ne converge pas exponentiellement. Cela arrive probablement à cause de la perte de précision lorsqu'on utilise les estimateurs $\pi_N^L$ et $d_{VT}^{20,L}$ au lieu des paramètres théoriques du problème : les puissances $(1-K^{-1})^{N_{j+1}-N_{j}}$ donnent des nombres très petits, et donc le test de converge exponentielle est extrêmement sensible aux distances $d_{VT}(\pi_N^L, \pi)$, qui varient selon $L$. Par conséquent, en prennant $N, L$ plus grands, on obtiendrait un résultat plus proche de la prévision théorique, mais le calcul exigerait plus de puissance computationnelle.

### S3)

Lorsque la loi cible change, on définit une nouvelle fonction :

```python
def func_sin(n):
    return math.fabs(math.sin(n))/math.factorial(n)

def func_sin_norm(n):
    Z = func_sin(1)
    stop_crit = math.pow(10, -9)
    i = 1
    while(func_sin(i+1)/Z-1 > stop_crit):
        i = i + 1
        Z = Z + func_sin(i)
    return func_sin(n)/Z
```

```python
def acceptation_s3(x, y, lambda_p):
    return function_poisson(x, lambda_p)*func_sin(y)/(function_poisson(y, lambda_p)*func_sin(x))
```

```python
def metropolis_s3(N, lambda_p):
    x = np.zeros(N+1, dtype=int)
    y = int(npr.poisson(lambda_p))
    x[0] = y
    
    for i in range(N):
            y = npr.poisson(lambda_p)
            u = np.random.uniform()
            #If acceptation > 1, then u will surely be smaller than acceptation. 
            if u < acceptation_s3(x[i-1],y,lambda_p):
                x[i+1] = y
            else:
                x[i+1] = x[i]
    return x
```

```python
def xn_estimator_s3(N, lambda_p, n_sample):
    xn = np.zeros(n_sample, dtype = int)
    for i in range(n_sample):
        xn[i] = metropolis_s3(N, lambda_p)[N-1]
    return xn
```

```python
lambda_p = 1
N = [5, 50, 100, 1000]
xtest = []
for i in range(len(N)):
    xtest.append(xn_estimator_s3(N[i], lambda_p, n_sample));

fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize=(16,12))
for i, ax in zip(range(len(xtest)), axs.flat):
    hist_plot(xtest[i], N[i], n_sample, lambda_p, ax)
```

On peut voir l'erreur de division par zéro, qui arrive lorsqu'on calcule la régle d'acceptation $R(0,y), \forall y \in \mathbb{N}$. Par conséquent, l'algorithme ne marche pas pour ce choix de $\pi$ et de $q$, vu que l'hypothèse $\pi(x) > 0,  \forall x \in E$ n'est plus satisfaite.
